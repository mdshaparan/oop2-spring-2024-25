# Simple  Relex Agent

```python
def simple_reflex_agent(percept):
  print("Perception Received: ",percept)
  loc = percept[0]
  status = percept[1]
  if status == 'Dirty' :
    action = 'Suck'
  elif loc == 'A' :
      action= 'Right'
  elif loc == 'B':
        action = 'Left'
        return action
```

```python
#Simulation

Location = random.choice(['A', 'B'])
Condition = random.choice(['Clean', 'Dirty'])

while True:
    action= simple_reflex_agent((Location,Condition))
    print('Action Performed: ', action)
    read= input('Continue reading? y/n: ')
    if (read == 'n' or read != 'y'):
     break
    if action == 'Right':
       Location= 'B'
       Condition= random.choice(['Clean','Dirty'])
    elif action == 'Left':
       Location = 'A'
       Condition= random.choice(['Clean','Dirty'])
    else:

       Condition= 'Clean'
```

# Model Based Agent

```python
import random

model = {'A':'Unknown', 'B':'Unknown'}
world_state = 'Bad'
action = 'Unknown'

def update_status(action, loc, status, model):
    model[loc] = status
    global world_state
    # update the model based on the action
    if action == 'Suck':
        model[loc] = 'Clean'
        # Check World
        if model['A'] == 'Clean' and model['B'] == 'Clean':
            world_state = 'Good'
        else:
            world_state = 'Bad'
    return world_state  # Always return world_state

def model_based_reflex_agent(percept):
    loc = percept[0]
    status = percept[1]
    global world_state, action, model

    if world_state == 'Good':
        action = 'Stop'
        # Print final status before stopping
        print('Perception: ', percept)
        print('Action Performed: ', action)
        print('Model: ', model)
        print('World: ', world_state)
        return action
    elif status == 'Dirty':
        action = 'Suck'
    elif loc == 'A':
        action = 'Right'
    elif loc == 'B':
        action = 'Left'

    world_state = update_status(action, loc, status, model)
    print('Perception: ', percept)
    print('Action Performed: ', action)
    print('Model: ', model)
    print('World: ', world_state)
    return action
```

```python
#simulation
Location = random.choice(['A', 'B'])
Condition = random.choice(['Clean', 'Dirty'])

while True:
    print('-------------')
    action = model_based_reflex_agent((Location, Condition))

    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['Clean','Dirty'])
    elif action == 'Left':
        Location = 'A'
        Condition = random.choice(['Clean','Dirty'])
    elif action == 'Suck':
        Condition = 'Clean'
    elif action == 'Stop':
        read = input('Job Done!. Do you want to Restart? y/n: ')
        if read == 'y':
            Location = random.choice(['A', 'B'])
            Condition = random.choice(['Clean', 'Dirty'])
            model = {'A':'Unknown', 'B':'Unknown'}
            world_state = 'Bad'
            action = 'Unknown'
        else:
            break
```
